# Simple Math Fuc^n in Python

## Linear Fuc^n
A linear function is a polynomial function of the first degree. Its graph is a straight line. The general form of a linear function is:

y = mx + b

Where:

    y is the dependent variable.

    x is the independent variable.

    m is the slope of the line, representing the rate of change of y with respect to x.

    b is the y-intercept, which is the value of y when x is 0.

```python
# Example  Y = 2 * X
x = 1
y = 2 * x
print(f"when x = {x}, y = {y}")
```

```python
# Example
x = 5
y = 2 * x
print(f"when x = {x}, Then Y = {y}")
```

```python
# How To Plot This Fuc^n
x_val = np.linspace(0,10,100)
y_val = 4 * x_val             # <-- Change The Value

plt.figure(figsize=(8,4))
plt.plot(x_val, y_val, color='green', linewidth=2, label='Y = 4x')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Fuc^n")
plt.grid(True)
plt.legend()
plt.show()
```

## Quadratic Fuc^n
A quadratic function is a polynomial function of the second degree. Its graph is a parabola (a U-shaped curve). The general form of a quadratic function is:

y = ax² + bx + c

Where:

    y is the dependent variable.
    x is the independent variable.
    a, b, and c are constants, with a not equal to 0.
    If a > 0, the parabola opens upwards.
    If a < 0, the parabola opens downwards.

```python
Example = [-2,0,1,2,3]
for X in Example:
  Y = X ** 2
  print(f"When X = {X}, Then Y = {Y}")

# Main Ploting Fuc^n
x_val = np.linspace(-5,5,100)
y_val = x_val ** 2

plt.figure(figsize=(8,4))
plt.plot(x_val, y_val, color='green', linewidth=2, label='Y = x^2')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Quadratic Fuc^n")
plt.grid(True)
plt.legend()
plt.show()
```

Understanding the Slope of a Quadratic Function

A quadratic function has an \(x^2\) term, which means when we square any number—whether positive or negative—the result is always positive. That’s why the **square part** of the function always gives non-negative values.  

However, this doesn’t mean the entire quadratic or its slope is always positive.  
For example, in \( f(x) = x^2 \), the output values are always positive or zero, but the **slope** (derivative) \( f'(x) = 2x \) changes based on \(x\):  
- When \(x\) is negative → slope is **negative**  
- When \(x\) is zero → slope is **zero**  
- When \(x\) is positive → slope is **positive**  

So, while squaring makes the function’s values positive, the **slope** itself is **not always positive** — it depends on where you are on the curve.


## Cubic Fuc^n

Understanding the Slope of a Cubic Function

A cubic function has an **x³** term, which means its shape and slope behave differently from a quadratic.

When you cube a number:
- A positive number stays **positive**
- A negative number stays **negative**

So, unlike squaring, cubing **keeps the sign** of the input.

Example
**f(x) = x³**

- For negative x → output is **negative**  
- For positive x → output is **positive**  
- At x = 0 → output is **zero**

The **slope** (derivative) is:

$$
f'(x) = 3x^2
$$

Since \(x^2\) is always non-negative, \(3x^2\) is always **positive or zero**.  
That means the slope of a simple cubic like \(x^3\) is **always positive**, except at \(x = 0\) where it’s zero.

However, for a general cubic:

$$
f(x) = ax^3 + bx^2 + cx + d
$$

the slope becomes:

$$
f'(x) = 3ax^2 + 2bx + c
$$

This can be **positive**, **negative**, or **zero**, depending on the coefficients and the value of \(x\).

---

**In short:**  
Cubic functions keep the sign of \(x\) (negative stays negative, positive stays positive),  
but their slope can change direction depending on the coefficients — it’s **not always positive**.


```python
Example = [-2,0,1,2,3]
for X in Example:
  Y = X ** 3
  print(f"When X = {X}, Then Y = {Y}")

# Main Ploting Fuc^n
x_val = np.linspace(-5,5,100)
y_val = x_val ** 3

plt.figure(figsize=(8,4))
plt.plot(x_val, y_val, color='green', linewidth=2, label='Y = x^3')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Cubic Fuc^n")
plt.grid(True)
plt.legend()
plt.show()
```

## Square Root Fuc^n
###  Understanding the Slope of a Square Root Function

A **square root function** has the form:

$$
f(x) = \sqrt{x}
$$

This function is defined **only for x ≥ 0**, because the square root of a negative number is not a real value.

#### Behavior of the Function
- For \(x > 0\), the output is **positive**.  
- For \(x = 0\), the output is **zero**.  
- As \(x\) increases, the growth of \(\sqrt{x}\) becomes **slower** — the curve rises quickly at first and then flattens out.

#### The Slope (Derivative)
The slope of the square root function is given by:

$$
f'(x) = \frac{1}{2\sqrt{x}}
$$

#### What This Means:
- For \(x > 0\), the slope is **positive**.  
- As \(x\) gets larger, \(\sqrt{x}\) increases slowly, so the slope **gets smaller**.  
- At \(x = 0\), the slope becomes **infinite** (very steep) because you are dividing by zero.  

---

**In short:**  
> The square root function is always **increasing** (slope positive),  
> but its slope keeps **decreasing** as x grows.  
> The curve starts very steep near 0 and flattens out for large x.


```python
Example = [-2,0,1,2,3]
for X in Example:
  Y = np.sqrt(X)
  print(f"When X = {X}, Then Y = {Y}")

# Main Ploting Fuc^n
x_val = np.linspace(-5,5,100)
y_val = np.sqrt(x_val)

plt.figure(figsize=(8,4))
plt.plot(x_val, y_val, color='green', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Square Root Fuc^n")
plt.grid(True)
plt.legend()
plt.show()
```

## Exponential Fuc^n
###  Understanding the Slope of an Exponential Function

An **exponential function** has the general form:

$$
f(x) = a^x
$$

where \(a > 0\) and \(a \neq 1\).

---

#### Behavior of the Function
- If \(a > 1\): the function **increases** rapidly as \(x\) increases.  
- If \(0 < a < 1\): the function **decreases** as \(x\) increases.  
- The value of \(f(x)\) is **always positive** — exponential functions never cross or touch the x-axis.

---

####  The Slope (Derivative)
The slope of the exponential function is:

$$
f'(x) = a^x \ln(a)
$$

---

####  Interpretation
- When \(a > 1\):  
  \( \ln(a) > 0 \), so \( f'(x) > 0 \) → the slope is **always positive**, meaning the function keeps **increasing**.
- When \(0 < a < 1\):  
  \( \ln(a) < 0 \), so \( f'(x) < 0 \) → the slope is **always negative**, meaning the function keeps **decreasing**.

---

####  Example
For the natural exponential function:

$$
f(x) = e^x
$$

The slope is:

$$
f'(x) = e^x
$$

Here, the derivative equals the function itself — both always positive, and the curve grows **exponentially faster** as \(x\) increases.

---

**In short:**  
> The exponential function is always **positive** in value.  
> If the base \(a > 1\), it **increases** with a **positive slope**.  
> If \(0 < a < 1\), it **decreases** with a **negative slope**.  
> For \(f(x) = e^x\), slope and value are the **same**, both growing infinitely as x increases.


```python
Example = [-2,0,1,2,3]
for X in Example:
  Y = 2 ** X
  print(f"When X = {X}, Then Y = 2^{X} = {Y}")

# Main Ploting Fuc^n
x_val = np.linspace(-5,5,100)
y_val = 2 ** x_val

plt.figure(figsize=(8,4))
plt.plot(x_val, y_val, color='green', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Exponential Fuc^n")
plt.grid(True)
plt.legend()
plt.show()
```

## Sine Fuc^n
### Understanding the Slope of a Sine Function

A sine function has the general form:

$$
f(x) = \sin(x)
$$

---

#### Behavior of the Function
The sine function oscillates between **–1** and **+1** and is **periodic** with a period of \(2\pi\).  
It crosses the x-axis at multiples of \(\pi\):  
\(x = 0, \pi, 2\pi, 3\pi, \dots\)

---

#### The Slope (Derivative)
The derivative of the sine function is:

$$
f'(x) = \cos(x)
$$

---

#### Interpretation
- When \(\cos(x) > 0\): the slope is **positive**, meaning the sine curve is rising.  
- When \(\cos(x) < 0\): the slope is **negative**, meaning the sine curve is falling.  
- When \(\cos(x) = 0\): the slope is **zero**, meaning the sine curve is flat at a peak or a trough.

Thus, the slope of the sine function changes continuously—it alternates between positive, zero, and negative as \(x\) increases.

---

#### Example
- At \(x = 0\): \(f(x) = 0\), \(f'(x) = 1\) → the curve is rising.  
- At \(x = \pi/2\): \(f(x) = 1\), \(f'(x) = 0\) → the curve is flat (maximum point).  
- At \(x = \pi\): \(f(x) = 0\), \(f'(x) = -1\) → the curve is falling.

---

**Summary:**  
The sine function oscillates smoothly between –1 and +1, while its slope (given by the cosine function) alternates between positive, zero, and negative values, producing the continuous wave-like pattern characteristic of trigonometric functions.


```python
Example = [-2,0,1,2,3]
for X in Example:
  Y = math.sin(X)                              #  <-- Change the sine overhere and experiment
  print(f"When X = {X:.2f}, Then Y = sin(X) = {Y:.2f}")

# Main Ploting Fuc^n
x_val = np.linspace(-5,5,100)
y_val = np.sin(x_val)

plt.figure(figsize=(8,4))
plt.plot(x_val, y_val, color='green', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sine Fuc^n")
plt.grid(True)
plt.legend()
plt.show()
```
