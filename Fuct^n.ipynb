## How Absolute values look like on graphs f(x) = |x| and see how it always going to give us results in the form of V shape

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Create interactive input fields
min_x_input = widgets.FloatText(description='Minimum x:', value=0)
max_x_input = widgets.FloatText(description='Maximum x:', value=0)
display(min_x_input, max_x_input)

def update_plot(change):
    min_x = min_x_input.value
    max_x = max_x_input.value

    x_abs = np.linspace(min_x, max_x, 400)
    y_abs = np.abs(x_abs)

    plt.figure(figsize=(8, 5))
    plt.plot(x_abs, y_abs, 'purple', label='|x|')
    plt.title("Graph of |x|")
    plt.xlabel("x")
    plt.ylabel("|x|")
    plt.grid(True)
    plt.legend()
    plt.show()

# Observe changes in input fields and update plot
min_x_input.observe(update_plot, names='value')
max_x_input.observe(update_plot, names='value')

# Initial plot
update_plot(None)
```

## Normal fuc^n

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the line
def line_eq(x):
    return x**3 - 6*x**2 + 9*x

# Generate x values
x_values = np.linspace(-10, 10, 400)
# Calculate corresponding y values
y_values = line_eq(x_values)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x_values, y_values, 'b-', label='$2x - 3y + 2 = 0$')
plt.title("Graph of the line $2x - 3y + 2 = 0$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

## Define Derivative put any fuc^n at define fuc^n and see how derivative can get tangent of that particular fuc^n

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return np.sin(x)

# Define derivative (approximate numerically)
def df(x):
    h = 1e-5
    return (f(x + h) - f(x - h)) / (2*h)

# Data range
x = np.linspace(-10, 10, 400)
y = f(x)

plt.figure(figsize=(10,6))
plt.plot(x, y, 'b', label='f(x)')

# Pick sample points to draw tangents
points = np.linspace(-8, 8, 9)
for p in points:
    slope = df(p)
    y_p = f(p)
    tangent_x = np.linspace(p-1.5, p+1.5, 20)
    tangent_y = y_p + slope * (tangent_x - p)
    plt.plot(tangent_x, tangent_y, 'r--', alpha=0.6)
    plt.plot(p, y_p, 'ko')

plt.title("Function with Tangent Lines (showing slope changes)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
